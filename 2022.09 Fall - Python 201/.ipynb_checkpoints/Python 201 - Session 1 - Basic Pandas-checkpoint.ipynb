```python
select_rooms = ["Club Deluxe King Room","King Suite"]
```

```python
select_cols = ["RoomType","RoomStats","April"]
```

```python
# when selecting rows & cols at the same time:
# loc - with labels
# iloc - with indices

plans.loc[plans.RoomType.isin(select_rooms),select_cols]
```
