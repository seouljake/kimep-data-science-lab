#### 6. sort and save the cities dataframe by admin_name, and then by population

- Sort cities first by two critiera:
 - by admin_name from A to Z
 - by population, from largest to smallest
- Save the sorted dataframe

#### 7. count and report number of cities by region

- Use the groupby function and an aggregate function to count the number of cities by each region of admin_name

```python
city_counts = cities.groupby(["admin_name"])["city"].count()
```

```python
city_counts
```

#### 8. count the cities and sum the population of the cities in each region

- Use the groupby function on admin_name
- Use the agg function with a dictionary in order to:
 - "count" city
 - "sum" population
- Save the output as a new dataframe, region_stats.
- Make sure that admin_name is also a variable and not an index. (Hint: Use .reset_index() to turn admin_name into a variable.)

Reference
- https://pandas.pydata.org/docs/reference/api/pandas.core.groupby.DataFrameGroupBy.aggregate.html

```python
region_stats = cities.groupby(["admin_name"]).agg({"city":"count","population":"sum"}).reset_index()
```

```python
region_stats
```

####  9. save and sum the populations of all the cities north of Qaraghandy

- Use the variable lat (for latitude) to select the number of cities north of Qaraghandy.
- Save the selected rows as a new dataframe called northern_cities.
- Use the variable population to sum the population of all of these cities.

```python
northern_cities = cities[cities.lat > 49.8000]
```

```python
northern_cities
```

```python
northern_cities["population"].sum()
```

####  10. merge regional_stats into the cities dataframe

- Merge region_stats (from Problem 8) to cities in order to add the regional statistics (city counts and population sum) of each city's region

Specifics
- cities should be the "left" dataframe and region_stats the "right" dataframe
- merge on "admin_name"
- using a left join
- apply the suffix "_inregion" to variables in admin_name
- don't apply a suffix to variables in cities

Reference
- https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.merge.html

```python
cities_with_region_stats = cities.merge(region_stats,on="admin_name",how="left",suffixes=("","_inregion"))
```

```python
cities_with_region_stats
```
