```python
pd.set_option('display.max_rows', None)
dfcounts = dfconcerns.groupby(["Room type","Concern description","Month"],as_index=False).size()
dfcounts
```

# construct time series (including zeroes)

```python
concerns = pd.DataFrame(list(dfcounts["Concern description"].unique()),columns=["Concern description"])
```

```python
concerns
```

```python
roomtypes = pd.DataFrame(list(dfcounts["Room type"].unique()),columns=["Room type"])
```

```python
roomtypes
```

```python
dfcounts["Month"].sort_values().unique()
```

```python
months = pd.DataFrame(list(dfcounts["Month"].sort_values().unique()),columns=["Month"])
```

```python
months
```

```python
ts1 = months.merge(concerns,how="cross")
```

```python
ts2 = roomtypes.merge(ts1,how="cross")
```

```python
ts2
```

```python
ts2.shape
```

```python
ts2["Count"]=0
```

```python
ts2
```

```python
dfcounts
```

```python
ts = ts2.merge(dfcounts,how="left",on=["Room type","Month","Concern description"])
```

```python
ts.shape
```

```python
ts
```

```python
ts.loc[ts["size"].notnull(),"Count"]=ts["size"]
```

```python
ts = ts.drop(columns="size")
ts["Count"] = ts["Count"].astype(int)
```

```python
ts
```

```python
ts.to_excel("output/concern time series.xlsx",index=False)
```

# time series - top three concerns for each room type

```python
ts
```

```python
dfconcernsbyroom = ts.groupby(["Room type","Concern description"],as_index=False)["Count"].agg(sum)
dfconcernsbyroom
```

```python
top3concerns = dfconcernsbyroom[dfconcernsbyroom.groupby(["Room type"])["Count"].rank(ascending=False)<=3]
top3concerns
```

```python
top3 = ts.merge(top3concerns,on=["Room type","Concern description"],how="inner")
top3.sort_values(["Room type","Month","Count_y","Concern description"],ascending=[True,True,False,True],inplace=True)
top3
```

```python
top3["ProblemNumber"] = [1,2,3] * 72
```

```python
# df.pivot(index='patient', columns='obs', values='score')

top3b = top3.pivot(index=["Room type","Month"],columns="ProblemNumber",values=["Concern description","Count_x"])
top3b.reset_index(inplace=True)
#top3b.columns.set_levels(["Room type", "Month", "Concern 1","Concern 2","Concern 3","Concern 1 Count","Concern 2 Count","Concern 3 Count"],level=1,inplace=True)
top3b.columns = [' '.join([str(a) for a in col]).strip() for col in top3b.columns.values]
top3b
```

```python
top3b
```

```python
top3b.to_excel("output/top 3 by room type and month.xlsx",index=False)
```
