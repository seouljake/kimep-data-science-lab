# Step 2. Re-calculate token frequencies

## We should have saved these at the end of Task 3. Oops!

reference: http://damir.cavar.me/pynotebooks/Python_NLTK_Texts_and_Frequencies.html

```python
# pip install nltk
```

```python
import nltk
from nltk import FreqDist
```

```python
# define fxn to generate the frequency distribution of a doc

def tokenfreqdist(tokens):

    freqdist = nltk.FreqDist(tokens)
    freqdist_tuples = [(x,freqdist[x]) for x in freqdist]
    freqdist_tuples.sort(key = lambda x: (-x[1],x[0]),reverse=False)
    return(freqdist_tuples)
```

```python
freqdist_tuples1 = tokenfreqdist(tokens1)
freqdist_tuples1
```

```python
freqdist_tuples2 = tokenfreqdist(tokens2)
freqdist_tuples2
```

# Step 3. Calculate Top 25 Tokens and Identify Overlapping Tokens

```python
top25_1 = freqdist_tuples1[0:25]
top25_2 = freqdist_tuples2[0:25]
```

```python
top25_tokens1 = [x[0] for x in top25_1]
top25_counts1 = [x[1] for x in top25_1]
```

```python
top25_tokens2 = [x[0] for x in top25_2]
top25_counts2 = [x[1] for x in top25_2]
```

```python
overlap = [x for x in top25_1 if x[0] in top25_tokens2]
len(overlap)
```

```python
overlap
```

```python
overlap_tokens = [x[0] for x in overlap]
overlap_counts = [x[1] for x in overlap]
overlap_tokens,overlap_counts
```

```python
top25_tokens1_only = [x[0] for x in top25_1 if x[0] not in overlap_tokens]
top25_counts1_only = [x[1] for x in top25_1 if x[0] not in overlap_tokens]
len(top25_tokens1_only)
```

```python
top25_tokens1_only,top25_counts1_only
```

```python
top25_tokens2_only = [x[0] for x in top25_2 if x[0] not in overlap_tokens]
top25_counts2_only = [x[1] for x in top25_2 if x[0] not in overlap_tokens]
len(top25_tokens2_only)
```

```python
top25_tokens2_only,top25_counts2_only
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize = (10, 5))
```

```python
# bar plot for doc1
bar_dat = dat1_all
bar_name = doc1name
bar_color = "red"

plt.bar(bar_dat["word"], bar_dat["weight"], color = bar_dat["color"], width = 0.4)
 
plt.xlabel("Tokens")
plt.xticks(rotation=90)
plt.ylabel("Frequency of Tokens")
plt.title("Top 25 Tokens in "+bar_name.capitalize())

colors = {doc1name+" and "+doc2name:'purple',bar_name+" only":bar_color}         
labels = list(colors.keys())
handles = [plt.Rectangle((0,0),1,1, color=colors[label]) for label in labels]
plt.legend(handles, labels)

# save the plot before showing
plt.savefig("../results/task4_top25_barplot_"+bar_name+".png",dpi=400)

plt.show()
```
