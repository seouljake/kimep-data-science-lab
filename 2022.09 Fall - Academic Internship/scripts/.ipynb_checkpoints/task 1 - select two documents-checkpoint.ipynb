# Document 1 - Hamlet

```python
# hamlet
gutenberg.words('shakespeare-hamlet.txt')
```

```python
# view original text
hamlet_text = " ".join(gutenberg.words('shakespeare-hamlet.txt'))
hamlet_text
```

```python
# export text
fileout = open("../data/hamlet_text.txt","w")
```

```python
#for item in hamlet_text:
fileout.write(hamlet_text)
```

```python
fileout.close()
```

```python
# words
hamlet_words = gutenberg.words('shakespeare-hamlet.txt')
type(hamlet_words)
```

```python
# How many words in the text:
print("Word count:", len(hamlet_words))
```

```python
hamlet_words
```

```python
hamlet_words[1000:2000]
```

```python
# convert nltk corpus reader object to a list
hamlet_tokens = list(hamlet_words)
len(hamlet_tokens)
```

```python
hamlet_tokens[0:100]
```

```python
type(hamlet_tokens)
```

```python
# export token list
fileout = open("../data/hamlet_tokens.txt","w")
```

```python
for item in hamlet_tokens:
    fileout.write(item+"\n")
```

```python
fileout.close()
```
