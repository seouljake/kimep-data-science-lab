#### 2. Generate Keyword Counts

```python
s = "Apples are delicious! I love apples!"
```

```python
fdict = {"apple":0,"orange":0,"grape":0,"melon":0,"pear":0}
```

```python
def countfruit(s,fdict):
    
    for f in fdict.keys():
        fdict[f] += s.lower().count(f)

    return fdict
```

```python
countfruit(s,fdict)
```

```python
def count_fruit_in_text(text):

    fruitcount = {"apple":0,"orange":0,"grape":0,"melon":0,"pear":0}
    fruitlist = list(fruitcount.keys())
    fruitsentence = []

    for line in text:
        
        # 1.updating fruit count
        fruitcount = countfruit(line,fruitcount)
        
        #2. check if this sentence contains a fruit keyword; and if so, append to a list
        for f in fruitlist:
            if f in line:
                fruitsentence.append(line)

    return fruitcount,fruitsentence
```

```python
fc,fs = count_fruit_in_text(text)
```
